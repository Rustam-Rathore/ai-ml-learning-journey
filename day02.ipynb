## 2. Loops – Automating Repetitive Tasks

```python
# for i in range(4):
#     print("Learning Python for AI & ML")

for i in range (1,10,3):
  print("iteration number:",i)

# print("iteration number:1")
```

```python
cgpa_list = [6.5, 7.2, 8.1, 5.9, 7.8]

for cgpa in cgpa_list:
    if cgpa >= 7:
        print(cgpa, "→ High placement chance")
    else:
        print(cgpa, "→ Needs improvement")
print(cgpa_list)
# range(start,stop,step)
```

## 3. Functions – Reusable Logic

```python
def placement_prediction(cgpa):
    if cgpa >= 7:
        return "Good placement chances"
    else:
        return "Needs improvement"
```

```python
print(placement_prediction(6.4))
print(placement_prediction(8.2))
```

## 4. Lists – Storing Multiple Values

```python
marks = [65, 72, 80, 90,"sid",7.2,True, 85]
print(marks)
print(len(marks))
```

```python
marks = [65, 72, 80, 90, 85]
marks.sort()
print(marks)
```

## 5. Indexing (Access Single Element)

```python
print("First element:", marks[0])
print("Third element:", marks[2])
print("Last element:", marks[-2])
print("Last element:", marks[4])
```

## 6. Slicing (Access Multiple Elements)

```python
print("First three marks:", marks[:3])
print("Middle marks:", marks[1:4])
print("Last two marks:", marks[-2:])
```

## 7. Dictionary – Structured Data

```python
student = {
    "name": "Amit",
    "cgpa": 7.6,
    "coding_skill": 3,
    "internships": 1
}

print(student)
print("CGPA:", student["cgpa"])
```

## 8. Arrays – Introduction to NumPy (AI Ready)

Arrays are used heavily in AI & ML because they are faster and more efficient than lists.

```python
import numpy as np

cgpa_array = np.array([6.5, 7.2, 8.1, 5.9, 7.8])
print(cgpa_array)
```

### Array Indexing & Slicing

```python
print("First CGPA:", cgpa_array[0])
print("Top three CGPAs:", cgpa_array[:3])
```

## 9. Visualization – Understanding Patterns

```python
import matplotlib.pyplot as plt

cgpa = [6, 7, 8, 9]
placement = [0, 1, 1, 1]

plt.bar(cgpa, placement)
plt.xlabel("CGPA")
plt.ylabel("Placement Status")
plt.title("CGPA vs Placement Outcome")
plt.show()
```

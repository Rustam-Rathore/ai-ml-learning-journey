## What is Machine Learning? (Real Meaning)

**Machine Learning means:**
- We do NOT write rules
- We give data
- The machine learns patterns from data

**Normal Programming:**
Rules + Data → Output

**Machine Learning:**
Data + Output → Rules (Model)


## AI vs ML vs Data Science

- **AI (Artificial Intelligence):** Making machines intelligent
- **ML (Machine Learning):** Machines learn from data
- **Data Science:** Understanding and analyzing data

**Memory Trick:**
- AI = Intelligence
- ML = Learning
- Data Science = Understanding Data


## Why NumPy is Required for Machine Learning

- ML models do not work with Python lists
- ML needs fast numerical computation
- NumPy provides optimized numeric arrays

```python
import numpy as np

cgpa = np.array([6.5, 7.0, 7.8, 8.2])
print(cgpa)
```

## Understanding NumPy Array Structure

```python
print("Shape of array:", cgpa.shape)
print("First CGPA:", cgpa[0])
print("Top three CGPAs:", cgpa[:3])
```

## Types of Machine Learning

### 1️. Supervised Learning
- Data + correct answer
- Example: Placement prediction

### 2️. Unsupervised Learning
- Only data, no answer
- Example: Student grouping

### 3️. Reinforcement Learning
- Learning using reward & punishment


## Student Placement Prediction – Project Start

- Input Feature: CGPA
- Output Target: Placement Status
- 0 = Not Placed, 1 = Placed


```python
placed = np.array([0, 1, 1, 1])

print("CGPA:", cgpa)
print("Placed:", placed)
```

## Visualization – Understanding Patterns Before ML


```python
import matplotlib.pyplot as plt

plt.plot(cgpa, placed, marker='o')
plt.xlabel("CGPA")
plt.title("CGPA vs Placement Outcome")
plt.ylabel("Placement Status")
plt.yticks([0, 1], ["Not Placed", "Placed"])

plt.show()
```
